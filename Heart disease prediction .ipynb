# Removing Outliers

```python
plt.figure(figsize=(12,4))
def drop_by_std(data,factor):
    upper_limit = data.cp.mean()+data.cp.std()*factor
    lower_limit = data.cp.mean()-data.cp.std()*factor
    x=data[(data.cp<upper_limit) & (data.cp>lower_limit)]
    return x
data_new = drop_by_std(data,2)
sns.boxplot(data_new['cp'])
```

```python
plt.figure(figsize=(12,4))
def drop_by_std(data,factor):
    upper_limit = data.trestbps.mean()+data.trestbps.std()*factor
    lower_limit = data.trestbps.mean()-data.trestbps.std()*factor
    x=data[(data.trestbps<upper_limit) & (data.trestbps>lower_limit)]
    return x
data_new = drop_by_std(data,2)
sns.boxplot(data_new['trestbps'])
```

# Applying Machine learining model

```python
from sklearn.preprocessing import StandardScaler
standardScaler = StandardScaler()
columns_to_scale = ['age', 'trestbps', 'chol', 'thalach', 'oldpeak']
data[columns_to_scale] = standardScaler.fit_transform(data[columns_to_scale])
```

```python
y = data['target']    #dependent variable
X = data.drop(['target'], axis = 1)   #independent variables
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test,y_train, y_test=train_test_split(X,y,test_size=0.3,random_state=40)
```

 # 1. Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
import pickle
lr=LogisticRegression()

model1=lr.fit(X_train,y_train)
pickle.dump(model1,open('model1.pkl','wb'))

model1= pickle.load(open('model1.pkl','rb'))
prediction1=model1.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(y_test,prediction1)
cm
```

```python
from sklearn import metrics
```

```python
metrics.accuracy_score(prediction1,y_test)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,prediction1)
```

```python
accuracy_lr=accuracy_score(y_test,prediction1)*100
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree


dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
```

```python
prediction=dt.predict(X_test)
accuracy_dt=accuracy_score(y_test,prediction)*100
```
